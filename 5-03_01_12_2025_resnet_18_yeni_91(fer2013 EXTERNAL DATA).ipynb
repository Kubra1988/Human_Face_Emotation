# Resnet full fine tune + Grad Cam

```python
import warnings
import os
import numpy as np
import pandas as pd
from pathlib import Path
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, models
from PIL import Image
from sklearn.model_selection import train_test_split
from sklearn.metrics import (classification_report, balanced_accuracy_score,
                            f1_score, accuracy_score, precision_score,
                            recall_score, confusion_matrix)
import matplotlib.pyplot as plt
import seaborn as sns
import json
import time
from tqdm import tqdm
import cv2

warnings.filterwarnings("ignore")

# ####################################################
#  VERI KLAS√ñR√ú
# ####################################################
path = '/content/drive/MyDrive/Data'
print(f"Dataset Path: {path}")

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Using device: {device}')


```

```python
# ####################################################
#  VERI KLAS√ñRLERI
# ####################################################
path = '/content/drive/MyDrive/Data'
fer2013_test_path = '/content/drive/MyDrive/test'  # FER2013 test klas√∂r√º
print(f"Dataset Path: {path}")
print(f"FER2013 Test Path: {fer2013_test_path}")

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Using device: {device}')
```

```python
# ----------------------------------------------------
# DATASET SINIFI
# ----------------------------------------------------
class EmotionDataset(Dataset):
    def __init__(self, image_paths, labels, transform=None):
        self.image_paths = image_paths
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        image = Image.open(img_path).convert('L')  # grayscale
        label = self.labels[idx]

        if self.transform:
            image = self.transform(image)

        return image, label, img_path


# ----------------------------------------------------
# DATASET Y√úKLEME
# ----------------------------------------------------
def load_dataset(data_path):
    image_paths = []
    labels = []
    class_names = sorted(os.listdir(data_path))
    class_to_idx = {class_name: idx for idx, class_name in enumerate(class_names)}

    for class_name in class_names:
        class_path = os.path.join(data_path, class_name)
        if os.path.isdir(class_path):
            for img_name in os.listdir(class_path):
                img_path = os.path.join(class_path, img_name)
                if img_path.lower().endswith(('.png', '.jpg', '.jpeg')):
                    image_paths.append(img_path)
                    labels.append(class_to_idx[class_name])

    return image_paths, labels, class_names

print("Loading dataset...")
image_paths, labels, class_names = load_dataset(path)
print(f"Total images: {len(image_paths)}")
print(f"Classes: {class_names}")


```

```python
# ----------------------------------------------------
# TRAIN / VAL / TEST SPLIT (70 / 10 / 20)
# ----------------------------------------------------
X_temp, X_test, y_temp, y_test = train_test_split(
    image_paths, labels, test_size=0.2, random_state=123, stratify=labels
)

X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.125, random_state=123, stratify=y_temp
)

print(f"Train size: {len(X_train)}, Val size: {len(X_val)}, Test size: {len(X_test)}")


# ----------------------------------------------------
# TRANSFORMS (224x224, GRAYSCALE)
# ----------------------------------------------------
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5], std=[0.5])
])


train_dataset = EmotionDataset(X_train, y_train, transform=transform)
val_dataset = EmotionDataset(X_val, y_val, transform=transform)
test_dataset = EmotionDataset(X_test, y_test, transform=transform)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True, num_workers=4)
val_loader = DataLoader(val_dataset, batch_size=64, shuffle=False, num_workers=4)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False, num_workers=4)


# ----------------------------------------------------
# MODEL - ResNet18 (FULL FINE-TUNE)
# ----------------------------------------------------
class EmotionResNet18(nn.Module):
    def __init__(self, num_classes=5):
        super(EmotionResNet18, self).__init__()

        # ResNet18 pretrained model y√ºkle
        self.resnet = models.resnet18(pretrained=True)

        # ƒ∞lk conv katmanƒ±nƒ± 1 kanallƒ± (grayscale) olarak deƒüi≈ütir
        self.resnet.conv1 = nn.Conv2d(1, 64, kernel_size=7, stride=2, padding=3, bias=False)

        # T√úM KATMANLAR Eƒûƒ∞Tƒ∞LECEK - Freeze yok!
        for param in self.resnet.parameters():
            param.requires_grad = True

        # Son fully connected katmanƒ±nƒ± deƒüi≈ütir
        in_features = self.resnet.fc.in_features
        self.resnet.fc = nn.Sequential(
            nn.Dropout(p=0.3),
            nn.Linear(in_features, num_classes)
        )

    def forward(self, x):
        return self.resnet(x)


model = EmotionResNet18(num_classes=len(class_names)).to(device)
print(model)


```

```python
# ----------------------------------------------------
# LOSS / OPTIMIZER
# ----------------------------------------------------
criterion = nn.CrossEntropyLoss()

# T√úM PARAMETRELERƒ∞ Eƒûƒ∞T
optimizer = optim.AdamW(model.parameters(), lr=1e-4, weight_decay=1e-4)  # LR d√º≈ü√ºr√ºld√º (full fine-tune i√ßin)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(
    optimizer, mode='min', factor=0.1, patience=3
)


# ----------------------------------------------------
# TRAIN + VAL FONKSƒ∞YONLARI
# ----------------------------------------------------
def train_epoch(model, dataloader, criterion, optimizer, device):
    model.train()
    running_loss, correct, total = 0, 0, 0

    for inputs, labels, _ in tqdm(dataloader, desc='Training', leave=False):
        inputs, labels = inputs.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)

        loss.backward()
        optimizer.step()

        running_loss += loss.item() * inputs.size(0)
        _, preds = outputs.max(1)
        total += labels.size(0)
        correct += preds.eq(labels).sum().item()

    return running_loss / total, correct / total


def validate_epoch(model, dataloader, criterion, device):
    model.eval()
    running_loss, correct, total = 0, 0, 0

    with torch.no_grad():
        for inputs, labels, _ in tqdm(dataloader, desc='Validation', leave=False):
            inputs, labels = inputs.to(device), labels.to(device)

            outputs = model(inputs)
            loss = criterion(outputs, labels)

            running_loss += loss.item() * inputs.size(0)
            _, preds = outputs.max(1)
            total += labels.size(0)
            correct += preds.eq(labels).sum().item()

    return running_loss / total, correct / total


# ----------------------------------------------------
# TRAINING LOOP
# ----------------------------------------------------
print("\nTraining started...")
num_epochs = 50
patience = 12
best_val_loss = float('inf')
patience_counter = 0

history = {
    'train_loss': [],
    'train_acc': [],
    'val_loss': [],
    'val_acc': []
}

for epoch in range(num_epochs):
    print(f"\nEpoch {epoch+1}/{num_epochs}")
    start_time = time.time()

    train_loss, train_acc = train_epoch(model, train_loader, criterion, optimizer, device)
    val_loss, val_acc = validate_epoch(model, val_loader, criterion, device)

    epoch_time = time.time() - start_time

    history['train_loss'].append(train_loss)
    history['train_acc'].append(train_acc)
    history['val_loss'].append(val_loss)
    history['val_acc'].append(val_acc)

    print(
        f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f} || "
        f"Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.4f} || "
        f"Epoch Time: {epoch_time:.2f} sec"
    )

    scheduler.step(val_loss)

    if val_loss < best_val_loss:
        best_val_loss = val_loss
        patience_counter = 0

        torch.save({
            'model_state_dict': model.state_dict(),
            'history': history,
            'class_names': class_names
        }, 'resnet18_full_finetune_emotion_model.pth')

        print(f"Best model saved (Val Loss: {val_loss:.4f})")
    else:
        patience_counter += 1
        if patience_counter >= patience:
            print(f"\nEarly stopping triggered at epoch {epoch+1}")
            break

```

```python
# ----------------------------------------------------
# TEST EVALUATION - Detaylƒ± Metrikler
# ----------------------------------------------------
def evaluate_test(model, dataloader, device):
    model.eval()
    all_labels, all_preds = [], []
    with torch.no_grad():
        for inputs, labels, _ in tqdm(dataloader, desc='Testing', leave=False):
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            _, pred = outputs.max(1)

            all_labels.extend(labels.cpu().numpy())
            all_preds.extend(pred.cpu().numpy())

    return np.array(all_labels), np.array(all_preds)


def calculate_sensitivity_specificity(y_true, y_pred, num_classes):
    """Her sƒ±nƒ±f i√ßin sensitivity ve specificity hesapla"""
    cm = confusion_matrix(y_true, y_pred)

    sensitivity = []
    specificity = []

    for i in range(num_classes):
        tp = cm[i, i]
        fn = cm[i, :].sum() - tp
        fp = cm[:, i].sum() - tp
        tn = cm.sum() - (tp + fn + fp)

        sens = tp / (tp + fn) if (tp + fn) > 0 else 0
        sensitivity.append(sens)

        spec = tn / (tn + fp) if (tn + fp) > 0 else 0
        specificity.append(spec)

    return sensitivity, specificity


print("\n" + "="*80)
print("TEST SET RESULTS")
print("="*80)

y_true, y_pred = evaluate_test(model, test_loader, device)

# Classification Report
print("\nüìä CLASSIFICATION REPORT:")
print("-"*80)
print(classification_report(y_true, y_pred, target_names=class_names, digits=4))

# Balanced Accuracy
bal_acc = balanced_accuracy_score(y_true, y_pred)
print(f"\nüéØ BALANCED ACCURACY: {bal_acc:.4f}")

# Class-wise Sensitivity ve Specificity
sensitivity, specificity = calculate_sensitivity_specificity(y_true, y_pred, len(class_names))

print("\nüìà CLASS-WISE METRICS:")
print("-"*80)
print(f"{'Class':<20} {'Sensitivity':<15} {'Specificity':<15}")
print("-"*80)
for i, class_name in enumerate(class_names):
    print(f"{class_name:<20} {sensitivity[i]:<15.4f} {specificity[i]:<15.4f}")

print("-"*80)
print(f"{'AVERAGE':<20} {np.mean(sensitivity):<15.4f} {np.mean(specificity):<15.4f}")
```

```python
# ============================================================
# GRAD-CAM IMPLEMENTATION
# ============================================================
print("\n" + "="*80)
print("GRAD-CAM VISUALIZATION")
print("="*80)

class GradCAM:
    def __init__(self, model, target_layer):
        self.model = model
        self.target_layer = target_layer
        self.gradients = None
        self.activations = None

        # Hook'larƒ± kaydet
        self.hook_layers()

    def hook_layers(self):
        def forward_hook(module, input, output):
            self.activations = output.detach()

        def backward_hook(module, grad_input, grad_output):
            self.gradients = grad_output[0].detach()

        self.target_layer.register_forward_hook(forward_hook)
        self.target_layer.register_full_backward_hook(backward_hook)

    def generate_cam(self, input_image, target_class=None):
        # Forward pass
        model_output = self.model(input_image)

        if target_class is None:
            target_class = model_output.argmax(dim=1)

        # Backward pass
        self.model.zero_grad()
        class_loss = model_output[0, target_class]
        class_loss.backward()

        # Grad-CAM hesapla
        gradients = self.gradients[0].cpu().numpy()
        activations = self.activations[0].cpu().numpy()

        # Global Average Pooling
        weights = np.mean(gradients, axis=(1, 2))

        # Weighted combination
        cam = np.zeros(activations.shape[1:], dtype=np.float32)
        for i, w in enumerate(weights):
            cam += w * activations[i]

        # ReLU
        cam = np.maximum(cam, 0)

        # Normalize
        if cam.max() > 0:
            cam = cam / cam.max()

        return cam, target_class.item()


def apply_colormap_on_image(org_im, activation, colormap=cv2.COLORMAP_JET):
    """Grad-CAM'i orijinal g√∂r√ºnt√ºye uygula"""
    heatmap = cv2.applyColorMap(np.uint8(255 * activation), colormap)
    heatmap = cv2.cvtColor(heatmap, cv2.COLOR_BGR2RGB)
    heatmap = np.float32(heatmap) / 255

    # Grayscale'i RGB'ye √ßevir
    if len(org_im.shape) == 2:
        org_im = np.stack([org_im] * 3, axis=-1)

    # Overlay
    cam = heatmap + np.float32(org_im)
    cam = cam / np.max(cam)

    return np.uint8(255 * cam)


def visualize_gradcam(model, test_loader, class_names, num_samples=10, device='cuda'):
    """Test setinden rastgele √∂rnekler se√ß ve Grad-CAM g√∂rselle≈ütir"""

    model.eval()

    # ResNet18'in son conv katmanƒ± (layer4)
    target_layer = model.resnet.layer4[-1].conv2

    gradcam = GradCAM(model, target_layer)

    # Rastgele √∂rnekler se√ß
    indices = np.random.choice(len(test_loader.dataset), num_samples, replace=False)

    fig, axes = plt.subplots(num_samples, 3, figsize=(15, 5*num_samples))

    for idx, sample_idx in enumerate(indices):
        img_tensor, true_label, img_path = test_loader.dataset[sample_idx]

        # Orijinal g√∂r√ºnt√º
        original_img = Image.open(img_path).convert('L')
        original_img_np = np.array(original_img.resize((224, 224))) / 255.0

        # Model prediction
        img_tensor = img_tensor.unsqueeze(0).to(device)

        with torch.no_grad():
            output = model(img_tensor)
            pred_class = output.argmax(dim=1).item()
            confidence = torch.softmax(output, dim=1)[0, pred_class].item()

        # Grad-CAM
        cam, _ = gradcam.generate_cam(img_tensor, target_class=pred_class)
        cam_resized = cv2.resize(cam, (224, 224))

        # Overlay
        visualization = apply_colormap_on_image(original_img_np, cam_resized)

        # Plot
        axes[idx, 0].imshow(original_img_np, cmap='gray')
        axes[idx, 0].set_title(f'Original\nTrue: {class_names[true_label]}')
        axes[idx, 0].axis('off')

        axes[idx, 1].imshow(cam_resized, cmap='jet')
        axes[idx, 1].set_title('Grad-CAM Heatmap')
        axes[idx, 1].axis('off')

        axes[idx, 2].imshow(visualization)
        axes[idx, 2].set_title(f'Overlay\nPred: {class_names[pred_class]}\nConf: {confidence:.2%}')
        axes[idx, 2].axis('off')

    plt.tight_layout()
    plt.savefig('gradcam_visualization.png', dpi=300, bbox_inches='tight')
    plt.show()

    print(f"\n‚úÖ Grad-CAM visualization saved: gradcam_visualization.png")


```

```python
import warnings
import os
import numpy as np
import pandas as pd
from pathlib import Path
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, models
from PIL import Image
from sklearn.model_selection import train_test_split
from sklearn.metrics import (classification_report, balanced_accuracy_score,
                            f1_score, accuracy_score, precision_score,
                            recall_score, confusion_matrix)
import matplotlib.pyplot as plt
import seaborn as sns
import json
import time
from tqdm import tqdm
import cv2

warnings.filterwarnings("ignore")

# ============================================================
# GRAD-CAM IMPLEMENTATION
# ============================================================
print("\n" + "="*80)
print("GRAD-CAM VISUALIZATION")
print("="*80)

class GradCAM:
    def __init__(self, model, target_layer):
        self.model = model
        self.target_layer = target_layer
        self.gradients = None
        self.activations = None

        # Hook'larƒ± kaydet
        self.hook_layers()

    def hook_layers(self):
        def forward_hook(module, input, output):
            self.activations = output.detach()

        def backward_hook(module, grad_input, grad_output):
            self.gradients = grad_output[0].detach()

        self.target_layer.register_forward_hook(forward_hook)
        self.target_layer.register_full_backward_hook(backward_hook)

    def generate_cam(self, input_image, target_class=None):
        # Forward pass
        model_output = self.model(input_image)

        if target_class is None:
            target_class = model_output.argmax(dim=1)

        # Backward pass
        self.model.zero_grad()
        # Ensure target_class is a tensor for indexing if it's passed as an int
        if isinstance(target_class, int):
            class_loss = model_output[0, target_class]
        else:
            class_loss = model_output[0, target_class]

        class_loss.backward()

        # Grad-CAM hesapla
        gradients = self.gradients[0].cpu().numpy()
        activations = self.activations[0].cpu().numpy()

        # Global Average Pooling
        weights = np.mean(gradients, axis=(1, 2))

        # Weighted combination
        cam = np.zeros(activations.shape[1:], dtype=np.float32)
        for i, w in enumerate(weights):
            cam += w * activations[i]

        # ReLU
        cam = np.maximum(cam, 0)

        # Normalize
        if cam.max() > 0:
            cam = cam / cam.max()

        # Return target_class directly as it might be an int or a tensor, and we want the value
        if isinstance(target_class, torch.Tensor):
            return cam, target_class.item()
        else:
            return cam, target_class


def apply_colormap_on_image(org_im, activation, colormap=cv2.COLORMAP_JET):
    """Grad-CAM'i orijinal g√∂r√ºnt√ºye uygula"""
    heatmap = cv2.applyColorMap(np.uint8(255 * activation), colormap)
    heatmap = cv2.cvtColor(heatmap, cv2.COLOR_BGR2RGB)
    heatmap = np.float32(heatmap) / 255

    # Grayscale'i RGB'ye √ßevir
    if len(org_im.shape) == 2:
        org_im = np.stack([org_im] * 3, axis=-1)

    # Overlay
    cam = heatmap + np.float32(org_im)
    cam = cam / np.max(cam)

    return np.uint8(255 * cam)


def visualize_gradcam(model, test_loader, class_names, num_samples=10, device='cuda'):
    """Test setinden rastgele √∂rnekler se√ß ve Grad-CAM g√∂rselle≈ütir"""

    model.eval()

    # ResNet18'in son conv katmanƒ± (layer4)
    target_layer = model.resnet.layer4[-1].conv2

    gradcam = GradCAM(model, target_layer)

    # Rastgele √∂rnekler se√ß
    indices = np.random.choice(len(test_loader.dataset), num_samples, replace=False)

    fig, axes = plt.subplots(num_samples, 3, figsize=(15, 5*num_samples))

    for idx, sample_idx in enumerate(indices):
        img_tensor, true_label, img_path = test_loader.dataset[sample_idx]

        # Orijinal g√∂r√ºnt√º
        original_img = Image.open(img_path).convert('L')
        original_img_np = np.array(original_img.resize((224, 224))) / 255.0

        # Model prediction
        img_tensor = img_tensor.unsqueeze(0).to(device)

        with torch.no_grad():
            output = model(img_tensor)
            pred_class = output.argmax(dim=1).item() # pred_class is an int here
            confidence = torch.softmax(output, dim=1)[0, pred_class].item()

        # Grad-CAM
        cam, _ = gradcam.generate_cam(img_tensor, target_class=pred_class) # pred_class is an int
        cam_resized = cv2.resize(cam, (224, 224))

        # Overlay
        visualization = apply_colormap_on_image(original_img_np, cam_resized)

        # Plot
        axes[idx, 0].imshow(original_img_np, cmap='gray')
        axes[idx, 0].set_title(f'Original\nTrue: {class_names[true_label]}')
        axes[idx, 0].axis('off')

        axes[idx, 1].imshow(cam_resized, cmap='jet')
        axes[idx, 1].set_title('Grad-CAM Heatmap')
        axes[idx, 1].axis('off')

        axes[idx, 2].imshow(visualization)
        axes[idx, 2].set_title(f'Overlay\nPred: {class_names[pred_class]}\nConf: {confidence:.2%}')
        axes[idx, 2].axis('off')

    plt.tight_layout()
    plt.savefig('gradcam_visualization.png', dpi=300, bbox_inches='tight')
    plt.show()

    print(f"\n‚úÖ Grad-CAM visualization saved: gradcam_visualization.png")
```

```python
import warnings
import os
import numpy as np
import pandas as pd
from pathlib import Path
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, models
from PIL import Image
from sklearn.model_selection import train_test_split
from sklearn.metrics import (classification_report, balanced_accuracy_score,
                            f1_score, accuracy_score, precision_score,
                            recall_score, confusion_matrix)
import matplotlib.pyplot as plt
import seaborn as sns
import json
import time
from tqdm import tqdm
import cv2

warnings.filterwarnings("ignore")

# ####################################################
# VERƒ∞ KLAS√ñR√ú
# ####################################################
path = '/content/drive/MyDrive/Data'
fer2013_test_path = '/content/drive/MyDrive/test'  # FER2013 test klas√∂r√º

print(f"Dataset Path: {path}")
print(f"FER2013 Test Path: {fer2013_test_path}")

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Using device: {device}')

# Model ve sonu√ßlar i√ßin Colab dizini
SAVE_DIR = '/content/'
MODEL_PATH = os.path.join(SAVE_DIR, 'resnet18_full_finetune_emotion_model.pth')
HISTORY_PATH = os.path.join(SAVE_DIR, 'training_history.json')

print(f"\nüìÅ Model save path: {MODEL_PATH}")
print(f"üìÅ History save path: {HISTORY_PATH}")

```

```python
# ----------------------------------------------------
# DATASET SINIFI
# ----------------------------------------------------
class EmotionDataset(Dataset):
    def __init__(self, image_paths, labels, transform=None):
        self.image_paths = image_paths
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        image = Image.open(img_path).convert('L')  # grayscale
        label = self.labels[idx]

        if self.transform:
            image = self.transform(image)

        return image, label, img_path

# ----------------------------------------------------
# DATASET Y√úKLEME
# ----------------------------------------------------
def load_dataset(data_path):
    image_paths = []
    labels = []
    class_names = sorted(os.listdir(data_path))
    class_to_idx = {class_name: idx for idx, class_name in enumerate(class_names)}

    for class_name in class_names:
        class_path = os.path.join(data_path, class_name)
        if os.path.isdir(class_path):
            for img_name in os.listdir(class_path):
                img_path = os.path.join(class_path, img_name)
                if img_path.lower().endswith(('.png', '.jpg', '.jpeg')):
                    image_paths.append(img_path)
                    labels.append(class_to_idx[class_name])

    return image_paths, labels, class_names

print("\nLoading dataset...")
image_paths, labels, class_names = load_dataset(path)
print(f"Total images: {len(image_paths)}")
print(f"Classes: {class_names}")

# ----------------------------------------------------
# TRAIN / VAL / TEST SPLIT (70 / 10 / 20)
# ----------------------------------------------------
X_temp, X_test, y_temp, y_test = train_test_split(
    image_paths, labels, test_size=0.2, random_state=123, stratify=labels
)
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.125, random_state=123, stratify=y_temp
)

print(f"Train size: {len(X_train)}, Val size: {len(X_val)}, Test size: {len(X_test)}")

# ----------------------------------------------------
# TRANSFORMS (224x224, GRAYSCALE)
# ----------------------------------------------------
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5], std=[0.5])
])

train_dataset = EmotionDataset(X_train, y_train, transform=transform)
val_dataset = EmotionDataset(X_val, y_val, transform=transform)
test_dataset = EmotionDataset(X_test, y_test, transform=transform)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True, num_workers=4)
val_loader = DataLoader(val_dataset, batch_size=64, shuffle=False, num_workers=4)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False, num_workers=4)

# FER2013 test set
if os.path.exists(fer2013_test_path):
    fer2013_image_paths, fer2013_labels, _ = load_dataset(fer2013_test_path)
    fer2013_dataset = EmotionDataset(fer2013_image_paths, fer2013_labels, transform=transform)
    fer2013_loader = DataLoader(fer2013_dataset, batch_size=64, shuffle=False, num_workers=4)
    print(f"FER2013 Test size: {len(fer2013_image_paths)}")
else:
    print(f"‚ö†Ô∏è  FER2013 test path not found: {fer2013_test_path}")
    fer2013_loader = None

# ----------------------------------------------------
# MODEL - ResNet18 (FULL FINE-TUNE)
# ----------------------------------------------------
class EmotionResNet18(nn.Module):
    def __init__(self, num_classes=5):
        super(EmotionResNet18, self).__init__()
        # ResNet18 pretrained model y√ºkle
        self.resnet = models.resnet18(pretrained=True)

        # ƒ∞lk conv katmanƒ±nƒ± 1 kanallƒ± (grayscale) olarak deƒüi≈ütir
        self.resnet.conv1 = nn.Conv2d(1, 64, kernel_size=7, stride=2, padding=3, bias=False)

        # T√úM KATMANLAR Eƒûƒ∞Tƒ∞LECEK - Freeze yok!
        for param in self.resnet.parameters():
            param.requires_grad = True

        # Son fully connected katmanƒ±nƒ± deƒüi≈ütir
        in_features = self.resnet.fc.in_features
        self.resnet.fc = nn.Sequential(
            nn.Dropout(p=0.3),
            nn.Linear(in_features, num_classes)
        )

    def forward(self, x):
        return self.resnet(x)

model = EmotionResNet18(num_classes=len(class_names)).to(device)
print("\n‚úÖ Model created successfully!")

# ----------------------------------------------------
# LOSS / OPTIMIZER
# ----------------------------------------------------
criterion = nn.CrossEntropyLoss()
optimizer = optim.AdamW(model.parameters(), lr=1e-4, weight_decay=1e-4)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(
    optimizer, mode='min', factor=0.1, patience=3
)

# ----------------------------------------------------
# TRAIN + VAL FONKSƒ∞YONLARI
# ----------------------------------------------------
def train_epoch(model, dataloader, criterion, optimizer, device):
    model.train()
    running_loss, correct, total = 0, 0, 0

    for inputs, labels, _ in tqdm(dataloader, desc='Training', leave=False):
        inputs, labels = inputs.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * inputs.size(0)
        _, preds = outputs.max(1)
        total += labels.size(0)
        correct += preds.eq(labels).sum().item()

    return running_loss / total, correct / total

def validate_epoch(model, dataloader, criterion, device):
    model.eval()
    running_loss, correct, total = 0, 0, 0

    with torch.no_grad():
        for inputs, labels, _ in tqdm(dataloader, desc='Validation', leave=False):
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)

            running_loss += loss.item() * inputs.size(0)
            _, preds = outputs.max(1)
            total += labels.size(0)
            correct += preds.eq(labels).sum().item()

    return running_loss / total, correct / total

# ----------------------------------------------------
# TRAINING LOOP
# ----------------------------------------------------
print("\n" + "="*80)
print("TRAINING STARTED")
print("="*80)

num_epochs = 50
patience = 6
best_val_loss = float('inf')
patience_counter = 0

history = {
    'train_loss': [],
    'train_acc': [],
    'val_loss': [],
    'val_acc': []
}

for epoch in range(num_epochs):
    print(f"\nEpoch {epoch+1}/{num_epochs}")
    start_time = time.time()

    train_loss, train_acc = train_epoch(model, train_loader, criterion, optimizer, device)
    val_loss, val_acc = validate_epoch(model, val_loader, criterion, device)

    epoch_time = time.time() - start_time

    history['train_loss'].append(train_loss)
    history['train_acc'].append(train_acc)
    history['val_loss'].append(val_loss)
    history['val_acc'].append(val_acc)

    print(
        f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f} || "
        f"Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.4f} || "
        f"Epoch Time: {epoch_time:.2f} sec"
    )

    scheduler.step(val_loss)

    if val_loss < best_val_loss:
        best_val_loss = val_loss
        patience_counter = 0

        # Model ve history'yi kaydet
        torch.save({
            'epoch': epoch + 1,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'history': history,
            'class_names': class_names,
            'best_val_loss': best_val_loss
        }, MODEL_PATH)

        # History'yi ayrƒ± JSON olarak kaydet
        with open(HISTORY_PATH, 'w') as f:
            json.dump(history, f, indent=4)

        print(f"‚úÖ Best model saved (Val Loss: {val_loss:.4f})")
    else:
        patience_counter += 1
        if patience_counter >= patience:
            print(f"\n‚èπÔ∏è  Early stopping triggered at epoch {epoch+1}")
            break

print("\n‚úÖ Training completed!")

# ----------------------------------------------------
# BEST MODEL Y√úKLE
# ----------------------------------------------------
print(f"\nüì• Loading best model from {MODEL_PATH}...")
checkpoint = torch.load(MODEL_PATH, map_location=device)
model.load_state_dict(checkpoint['model_state_dict'])
history = checkpoint['history']
print("‚úÖ Best model loaded successfully!")

# ----------------------------------------------------
# PLOT - Loss ve Accuracy Grafikleri
# ----------------------------------------------------
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

ax1.plot(history['train_loss'], label='Training Loss', linewidth=2)
ax1.plot(history['val_loss'], label='Validation Loss', linewidth=2)
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss')
ax1.set_title('Model Loss')
ax1.legend()
ax1.grid(True)

ax2.plot(history['train_acc'], label='Training Accuracy', linewidth=2)
ax2.plot(history['val_acc'], label='Validation Accuracy', linewidth=2)
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Accuracy')
ax2.set_title('Model Accuracy')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.savefig(os.path.join(SAVE_DIR, 'training_plots.png'), dpi=300)
plt.show()

# ============================================================
# EVALUATION FUNCTIONS
# ============================================================
def evaluate_test(model, dataloader, device):
    """Test seti √ºzerinde tahmin yap"""
    model.eval()
    all_labels, all_preds = [], []

    with torch.no_grad():
        for inputs, labels, _ in tqdm(dataloader, desc='Evaluating', leave=False):
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            _, pred = outputs.max(1)

            all_labels.extend(labels.cpu().numpy())
            all_preds.extend(pred.cpu().numpy())

    return np.array(all_labels), np.array(all_preds)

def calculate_sensitivity_specificity(y_true, y_pred, num_classes):
    """Her sƒ±nƒ±f i√ßin sensitivity ve specificity hesapla"""
    cm = confusion_matrix(y_true, y_pred)
    sensitivity = []
    specificity = []

    for i in range(num_classes):
        tp = cm[i, i]
        fn = cm[i, :].sum() - tp
        fp = cm[:, i].sum() - tp
        tn = cm.sum() - (tp + fn + fp)

        sens = tp / (tp + fn) if (tp + fn) > 0 else 0
        sensitivity.append(sens)

        spec = tn / (tn + fp) if (tn + fp) > 0 else 0
        specificity.append(spec)

    return sensitivity, specificity

def print_detailed_metrics(y_true, y_pred, class_names, dataset_name="Test Set"):
    """Detaylƒ± metrikleri yazdƒ±r"""
    print(f"\n{'='*80}")
    print(f"{dataset_name.upper()} RESULTS")
    print(f"{'='*80}")

    # Classification Report
    print("\nüìä CLASSIFICATION REPORT:")
    print("-"*80)
    print(classification_report(y_true, y_pred, target_names=class_names, digits=4))

    # Balanced Accuracy
    bal_acc = balanced_accuracy_score(y_true, y_pred)
    acc = accuracy_score(y_true, y_pred)
    print(f"\nüéØ ACCURACY: {acc:.4f}")
    print(f"üéØ BALANCED ACCURACY: {bal_acc:.4f}")

    # Class-wise Sensitivity ve Specificity
    sensitivity, specificity = calculate_sensitivity_specificity(y_true, y_pred, len(class_names))

    print("\nüìà CLASS-WISE METRICS:")
    print("-"*80)
    print(f"{'Class':<20} {'Sensitivity':<15} {'Specificity':<15}")
    print("-"*80)
    for i, class_name in enumerate(class_names):
        print(f"{class_name:<20} {sensitivity[i]:<15.4f} {specificity[i]:<15.4f}")
    print("-"*80)
    print(f"{'AVERAGE':<20} {np.mean(sensitivity):<15.4f} {np.mean(specificity):<15.4f}")

    return bal_acc, sensitivity, specificity

def plot_confusion_matrix(y_true, y_pred, class_names, dataset_name="Test Set"):
    """Confusion matrix √ßiz"""
    cm = confusion_matrix(y_true, y_pred)

    # Normal Confusion Matrix
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names,
                cbar_kws={'label': 'Count'})
    plt.title(f'Confusion Matrix - {dataset_name}', fontsize=16, fontweight='bold')
    plt.ylabel('True Label', fontsize=12)
    plt.xlabel('Predicted Label', fontsize=12)
    plt.tight_layout()

    safe_name = dataset_name.lower().replace(' ', '_')
    plt.savefig(os.path.join(SAVE_DIR, f'confusion_matrix_{safe_name}.png'), dpi=300)
    plt.show()

    # Normalized Confusion Matrix
    cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    plt.figure(figsize=(10, 8))
    sns.heatmap(cm_normalized, annot=True, fmt='.2%', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names,
                cbar_kws={'label': 'Percentage'})
    plt.title(f'Normalized Confusion Matrix - {dataset_name}', fontsize=16, fontweight='bold')
    plt.ylabel('True Label', fontsize=12)
    plt.xlabel('Predicted Label', fontsize=12)
    plt.tight_layout()
    plt.savefig(os.path.join(SAVE_DIR, f'confusion_matrix_normalized_{safe_name}.png'), dpi=300)
    plt.show()

    return cm

# ============================================================
# GRAD-CAM IMPLEMENTATION
# ============================================================
class GradCAM:
    def __init__(self, model, target_layer):
        self.model = model
        self.target_layer = target_layer
        self.gradients = None
        self.activations = None
        self.hook_layers()

    def hook_layers(self):
        def forward_hook(module, input, output):
            self.activations = output.detach()

        def backward_hook(module, grad_input, grad_output):
            self.gradients = grad_output[0].detach()

        self.target_layer.register_forward_hook(forward_hook)
        self.target_layer.register_full_backward_hook(backward_hook)

    def generate_cam(self, input_image, target_class=None):
        """Grad-CAM hesapla - D√úZELTILMI≈û VERSƒ∞YON"""
        # Forward pass
        model_output = self.model(input_image)

        if target_class is None:
            target_class = model_output.argmax(dim=1).item()

        # target_class zaten integer ise, direkt kullan
        if isinstance(target_class, int):
            class_idx = target_class
        else:
            class_idx = target_class.item()

        # Backward pass
        self.model.zero_grad()
        class_loss = model_output[0, class_idx]
        class_loss.backward()

        # Grad-CAM hesapla
        gradients = self.gradients[0].cpu().numpy()
        activations = self.activations[0].cpu().numpy()

        # Global Average Pooling
        weights = np.mean(gradients, axis=(1, 2))

        # Weighted combination
        cam = np.zeros(activations.shape[1:], dtype=np.float32)
        for i, w in enumerate(weights):
            cam += w * activations[i]

        # ReLU
        cam = np.maximum(cam, 0)

        # Normalize
        if cam.max() > 0:
            cam = cam / cam.max()

        return cam, class_idx

def apply_colormap_on_image(org_im, activation, colormap=cv2.COLORMAP_JET):
    """Grad-CAM'i orijinal g√∂r√ºnt√ºye uygula"""
    heatmap = cv2.applyColorMap(np.uint8(255 * activation), colormap)
    heatmap = cv2.cvtColor(heatmap, cv2.COLOR_BGR2RGB)
    heatmap = np.float32(heatmap) / 255

    # Grayscale'i RGB'ye √ßevir
    if len(org_im.shape) == 2:
        org_im = np.stack([org_im] * 3, axis=-1)

    # Overlay
    cam = heatmap + np.float32(org_im)
    cam = cam / np.max(cam)

    return np.uint8(255 * cam)

def visualize_gradcam(model, test_loader, class_names, dataset_name="Test Set",
                     num_samples=10, device='cuda'):
    """Test setinden rastgele √∂rnekler se√ß ve Grad-CAM g√∂rselle≈ütir"""
    model.eval()

    # ResNet18'in son conv katmanƒ± (layer4)
    target_layer = model.resnet.layer4[-1].conv2
    gradcam = GradCAM(model, target_layer)

    # Rastgele √∂rnekler se√ß
    indices = np.random.choice(len(test_loader.dataset), num_samples, replace=False)

    fig, axes = plt.subplots(num_samples, 3, figsize=(15, 5*num_samples))
    if num_samples == 1:
        axes = axes.reshape(1, -1)

    for idx, sample_idx in enumerate(indices):
        img_tensor, true_label, img_path = test_loader.dataset[sample_idx]

        # Orijinal g√∂r√ºnt√º
        original_img = Image.open(img_path).convert('L')
        original_img_np = np.array(original_img.resize((224, 224))) / 255.0

        # Model prediction
        img_tensor = img_tensor.unsqueeze(0).to(device)
        with torch.no_grad():
            output = model(img_tensor)
            pred_class = output.argmax(dim=1).item()
            confidence = torch.softmax(output, dim=1)[0, pred_class].item()

        # Grad-CAM
        cam, _ = gradcam.generate_cam(img_tensor, target_class=pred_class)
        cam_resized = cv2.resize(cam, (224, 224))

        # Overlay
        visualization = apply_colormap_on_image(original_img_np, cam_resized)

        # Plot
        axes[idx, 0].imshow(original_img_np, cmap='gray')
        axes[idx, 0].set_title(f'Original\nTrue: {class_names[true_label]}')
        axes[idx, 0].axis('off')

        axes[idx, 1].imshow(cam_resized, cmap='jet')
        axes[idx, 1].set_title('Grad-CAM Heatmap')
        axes[idx, 1].axis('off')

        axes[idx, 2].imshow(visualization)
        axes[idx, 2].set_title(f'Overlay\nPred: {class_names[pred_class]}\nConf: {confidence:.2%}')
        axes[idx, 2].axis('off')

    plt.tight_layout()
    safe_name = dataset_name.lower().replace(' ', '_')
    plt.savefig(os.path.join(SAVE_DIR, f'gradcam_{safe_name}.png'), dpi=300, bbox_inches='tight')
    plt.show()

    print(f"\n‚úÖ Grad-CAM visualization saved: gradcam_{safe_name}.png")

# ============================================================
# ORIGINAL TEST SET EVALUATION
# ============================================================
print("\n" + "="*80)
print("EVALUATING ON ORIGINAL TEST SET")
print("="*80)

y_true, y_pred = evaluate_test(model, test_loader, device)
bal_acc, sens, spec = print_detailed_metrics(y_true, y_pred, class_names, "Original Test Set")
cm = plot_confusion_matrix(y_true, y_pred, class_names, "Original Test Set")

# Grad-CAM visualization
visualize_gradcam(model, test_loader, class_names, "Original Test Set",
                 num_samples=10, device=device)

# Save results
results = {
    'dataset': 'Original Test Set',
    'balanced_accuracy': float(bal_acc),
    'accuracy': float(accuracy_score(y_true, y_pred)),
    'class_names': class_names,
    'sensitivity': [float(s) for s in sens],
    'specificity': [float(s) for s in spec],
    'confusion_matrix': cm.tolist()
}

with open(os.path.join(SAVE_DIR, 'test_results_original.json'), 'w') as f:
    json.dump(results, f, indent=4)

# ============================================================
# FER2013 TEST SET EVALUATION
# ============================================================
if fer2013_loader is not None:
    print("\n" + "="*80)
    print("EVALUATING ON FER2013 TEST SET")
    print("="*80)

    y_true_fer, y_pred_fer = evaluate_test(model, fer2013_loader, device)
    bal_acc_fer, sens_fer, spec_fer = print_detailed_metrics(
        y_true_fer, y_pred_fer, class_names, "FER2013 Test Set"
    )
    cm_fer = plot_confusion_matrix(y_true_fer, y_pred_fer, class_names, "FER2013 Test Set")

    # Grad-CAM visualization
    visualize_gradcam(model, fer2013_loader, class_names, "FER2013 Test Set",
                     num_samples=10, device=device)

    # Save results
    results_fer = {
        'dataset': 'FER2013 Test Set',
        'balanced_accuracy': float(bal_acc_fer),
        'accuracy': float(accuracy_score(y_true_fer, y_pred_fer)),
        'class_names': class_names,
        'sensitivity': [float(s) for s in sens_fer],
        'specificity': [float(s) for s in spec_fer],
        'confusion_matrix': cm_fer.tolist()
    }

    with open(os.path.join(SAVE_DIR, 'test_results_fer2013.json'), 'w') as f:
        json.dump(results_fer, f, indent=4)

# ============================================================
# FINAL SUMMARY
# ============================================================
print("\n" + "="*80)
print("üéâ ALL PROCESSING COMPLETED!")
print("="*80)

print("\nüìÅ Saved files in /content/:")
print("="*80)
print("\nüì¶ Model & Training:")
print(f"  ‚úì {MODEL_PATH}")
print(f"  ‚úì {HISTORY_PATH}")
print(f"  ‚úì training_plots.png")

print("\nüìä Original Test Set:")
print("  ‚úì test_results_original.json")
print("  ‚úì confusion_matrix_original_test_set.png")
print("  ‚úì confusion_matrix_normalized_original_test_set.png")
print("  ‚úì gradcam_original_test_set.png")

if fer2013_loader is not None:
    print("\nüìä FER2013 Test Set:")
    print("  ‚úì test_results_fer2013.json")
    print("  ‚úì confusion_matrix_fer2013_test_set.png")
    print("  ‚úì confusion_matrix_normalized_fer2013_test_set.png")
    print("  ‚úì gradcam_fer2013_test_set.png")

print("\n" + "="*80)
print("‚úÖ You can find all files in /content/ directory")
print("="*80)
```
